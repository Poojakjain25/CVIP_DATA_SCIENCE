PROJECT NAME : Exploratory Data Analysis - Terrorism

AIM : The aim of the project is to perform data analysis on the global terrorism dataset and understanding the characteristics, patterns, and underlying trends related to terrorist incidents.

WORKING : We use machine learning in order to understand deeper meaning and patterns related to terrorist activites due to its surge in recent years.
```python
# Importing Modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the csv file
terrorism_data = pd.read_csv("/content/globalterrorismdb_0718dist.csv", encoding = "ISO-8859-1")
```

There exists about 135 columns in the dataset, but only few important columns that contribute to analysis are choosen. In this about 20 out of 135 are choosen.

```python
terrorism_data_updated = terrorism_data[['eventid', 'iyear', 'imonth', 'iday', 'country_txt','region_txt', 'provstate', 'city', 'latitude', 'longitude','attacktype1_txt','nkill', 'nwound','target1', 'summary', 'gname','targtype1_txt','weaptype1_txt','motive','success']]
```

```python
terrorism_data_updated.shape
print("Number of Rows: ",terrorism_data_updated.shape[0])
print("Number of Columns: ",terrorism_data_updated.shape[1])
```

Determining the null values for each attribute in the dataset

```python
terrorism_data_updated.isnull().sum()
```

```python
print("The total number of null values in the entire dataset is :", terrorism_data_updated.isnull().sum().sum())
```

Determining the number of terrorist activities that took place from the year 1970 to 2017.

```python
plt.subplots(figsize=(20,10))
sns.countplot(data=terrorism_data_updated, x='iyear', palette='coolwarm')
plt.xticks(rotation=90)
plt.title('Terrorist Incidents from 1970 to 2017')
plt.show()
```

Determining the number of attacks for each region.

```python
attack_region = terrorism_data_updated.groupby(['iyear','region_txt']).size().reset_index(name='count')
plt.subplots(figsize=(20,10))
sns.lineplot(x='iyear',y='count',hue='region_txt',data=attack_region)
plt.title('Attack Trends Graph for each region')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.show()
```

Determining the top 10 countries that are affected by terrorist attacks.

```python
plt.subplots(figsize=(20,10))
top_countries=terrorism_data_updated['country_txt'].value_counts()[:10].to_frame().reset_index()
top_countries.columns= ['country_txt','attacktype1_tx']
sns.barplot(x='country_txt',y='attacktype1_tx', data= top_countries, palette='muted')
plt.title('Top 10 Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Attacks')
plt.xticks(rotation=20)
plt.show()
```

Determining the top 20 cities affected by terrorist attacks.

```python
plt.subplots(figsize=(20,10))
top_cities=terrorism_data_updated['city'].value_counts()[:20].to_frame().reset_index()
top_cities.columns= ['city','Attacks_Counts']
sns.barplot(x='city',y='Attacks_Counts', data= top_cities, palette='pastel')
plt.title('Top 20 affected cities')
plt.xlabel('city')
plt.ylabel('Attacks')
plt.xticks(rotation=80)
plt.show()
```

Determining top attacking methods used by terrorist.

```python
plt.subplots(figsize=(20,10))
sns.countplot(x='attacktype1_txt',data=terrorism_data_updated,palette='inferno',order=terrorism_data_updated['attacktype1_txt'].value_counts().index)
plt.xticks(rotation=80)
plt.title('Attacking Methods by Terrorists')
plt.show()
```

Determining the top 10 terrorist activity groups.

```python
group_counts = terrorism_data_updated['gname'].value_counts()
sort = group_counts.sort_values(ascending=False)
sort = sort.iloc[1:]
top_5 = sort.head(10)
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(20,10))
sns.barplot(x=top_5.index, y=top_5.values, palette="magma")
plt.title('Top 10 Terrorist Attack Groups')
plt.ylabel('Attack Count')
plt.xlabel('Group')
plt.xticks(rotation = 80)
plt.show()
```

Determining the activities by each terrorist attack groups over the years.

```python
top_10_groups=terrorism_data_updated[terrorism_data_updated['gname'].isin(terrorism_data_updated['gname'].value_counts()[1:11].index)]
pd.crosstab(top_10_groups.iyear,top_10_groups.gname).plot(color=sns.color_palette('Paired',10))
fig=plt.gcf()
fig.set_size_inches(20,10)
plt.show()
```

Determining the total number of people killed over the years due to terrorist attacks.

```python
killed_people=terrorism_data_updated[["iyear","nkill"]].groupby("iyear").sum()
plt.figure(figsize=(20,10))
sns.barplot(x=killed_people.index, y="nkill", palette="colorblind",data=killed_people)

plt.title("Total Killings by Year")
plt.xlabel("Year")
plt.ylabel("Total Killings")

plt.xticks(rotation=90)

plt.tight_layout()
plt.show()

k=terrorism_data_updated[["region_txt","nkill"]].groupby("region_txt").sum().sort_values(by="nkill",ascending=False)
k
```

Determining the number of people wounded over the years due to terrorist attacks.

```python
wounded_people=terrorism_data_updated[["iyear","nwound"]].groupby("iyear").sum()
plt.figure(figsize=(10, 6))
sns.barplot(x=wounded_people.index, y="nwound", palette="deep",data=wounded_people)

plt.title("Total Wounded by Year")
plt.xlabel("Year")
plt.ylabel("Total Wounded")

plt.xticks(rotation=90)

plt.tight_layout()
plt.show()

w=terrorism_data_updated[["region_txt","nwound"]].groupby("region_txt").sum().sort_values(by="nwound",ascending=False)
w
```

Analyzing data for India.

```python
India = terrorism_data_updated[(terrorism_data_updated['country_txt'] == 'India')]
India.head(5)
```

```python
India_attacks = India['eventid'].count()
print('There were',India_attacks ,'attacks in India.')
```

Determining the types of attacks that took place in India.

```python
attack_in_India = India.groupby(['attacktype1_txt','success']).size().reset_index(name='count')
attack_in_India
```

```python

plt.figure(figsize=(20,10))
plt.bar(attack_in_India['attacktype1_txt'], attack_in_India['count'], color='orange')
plt.xticks(rotation=90)
plt.title('Attacks by Type in India')
plt.xlabel('Attack Type')
plt.ylabel('Count')
plt.show()
```

Determining the number of people killed in India

```python
number_of_people_killed_India = India.groupby('attacktype1_txt')[['nkill']].sum().reset_index()
number_of_people_killed_India
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='attacktype1_txt', y='nkill', data=number_of_people_killed_India,palette= 'deep')
plt.xticks(rotation=90)
```
